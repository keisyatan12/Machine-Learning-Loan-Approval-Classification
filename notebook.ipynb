```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import joblib

# Define columns for encoding
categorical_columns = [
    'person_education', 
    'person_home_ownership', 
    'loan_intent', 
    'person_gender', 
    'previous_loan_defaults_on_file'
]

# Initialize OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, dtype=int, handle_unknown='ignore')

# Fit and transform the encoder on the categorical columns
encoded_data = encoder.fit_transform(df[categorical_columns])

# Create a DataFrame with the encoded column names and enforce integer dtype
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(categorical_columns), dtype=int, index=df.index)

# Combine with the original DataFrame (dropping the original categorical columns)
df_encoded = pd.concat([df.drop(columns=categorical_columns), encoded_df], axis=1)

# Save the encoder to use it in an API
joblib.dump(encoder, 'onehot_encoder.pkl')

# Display the encoded DataFrame
print(df_encoded.head())
```

```python
# Melakukan pengecekan missing values.
print("\nMissing Values in Each Column:")
print(df_encoded.isnull().sum())
```

```python
# Memisahkan kolom (independent variables) dan target (dependent variable) dalam dataset
x = df_encoded.drop(columns=['loan_status'])
y = df_encoded['loan_status']
```

```python
from sklearn.preprocessing import StandardScaler
import joblib
scaler = StandardScaler()

x_scaled = scaler.fit_transform(x)

x_scaled_df = pd.DataFrame(x_scaled, columns=x.columns)

joblib.dump(scaler, 'scaler.pkl')
x_scaled_df
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_resampled, y_resampled = smote.fit_resample(x_scaled_df, y)
print(pd.Series(y_resampled).value_counts())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, classification_report, confusion_matrix
import joblib
import numpy as np

# Define cross-validation strategy
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Define models to evaluate
models = {
    'Random Forest': RandomForestClassifier(random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    'Logistic Regression': LogisticRegression(random_state=42)
}

# Function to evaluate each model
def evaluate_model(model, x_resampled, y_resampled):
    precision = cross_val_score(model, x_resampled, y_resampled, cv=cv, scoring='precision')
    recall = cross_val_score(model, x_resampled, y_resampled, cv=cv, scoring='recall')
    f1 = cross_val_score(model, x_resampled, y_resampled, cv=cv, scoring='f1')
    roc_auc = cross_val_score(model, x_resampled, y_resampled, cv=cv, scoring='roc_auc')
    return {
        'Precision': (np.mean(precision), np.std(precision)),
        'Recall': (np.mean(recall), np.std(recall)),
        'F1 Score': (np.mean(f1), np.std(f1)),
        'ROC AUC': (np.mean(roc_auc), np.std(roc_auc))
    }

# Evaluate each model and store results
results = {}
for model_name, model in models.items():
    results[model_name] = evaluate_model(model, x_resampled, y_resampled)

# Print the comparison results
print("Cross-Validation Performance Comparison:")
for model_name, metrics in results.items():
    print(f"\n{model_name}:")
    for metric_name, (mean, std) in metrics.items():
        print(f"{metric_name}: {mean:.4f} Â± {std:.4f}")
```
